```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
X = data.drop('LUNG_CANCER', axis=1)
y = data['LUNG_CANCER']
```

```python
label_encoder = LabelEncoder()
X_encoded = X.apply(label_encoder.fit_transform)
```

```python
importance = mutual_info_classif(X_encoded, y)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(X.columns, importance)
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.title('Mutual Information for Feature Selection')
plt.show()
```

```python
feat_importance = pd.Series(importance, new_data.columns[0: len(new_data.columns)-1])
```

```python
feat_importance.plot(kind = 'barh', color = 'teal')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = new_data[['ALLERGY', 'COUGHING', 'ANXIETY']]
y = new_data['LUNG_CANCER']

regression_allergy = LinearRegression()
regression_coughing = LinearRegression()
# Fit the models to the data
regression_allergy.fit(X[['ALLERGY']], y)
regression_coughing.fit(X[['COUGHING']], y)

# Make predictions based on the models
y_pred_allergy = regression_allergy.predict(X[['ALLERGY']])
y_pred_coughing = regression_coughing.predict(X[['COUGHING']])

# Create separate scatter plots with regression lines for each feature
plt.figure(figsize=(10, 6))

# Scatter plot with regression line for ALLERGY
plt.scatter(X['ALLERGY'], y, label='ALLERGY Data', color='red')
plt.plot(X['ALLERGY'], y_pred_allergy, color='red', label='ALLERGY Regression Line')

# Scatter plot with regression line for COUGHING
plt.scatter(X['COUGHING'], y, label='COUGHING Data', color='blue')
plt.plot(X['COUGHING'], y_pred_coughing, color='blue', label='COUGHING Regression Line')



plt.xlabel('Features')
plt.ylabel('LUNG_CANCER')
plt.legend()
plt.show()
```
